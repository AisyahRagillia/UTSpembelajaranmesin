# Import Dataset

```python
import pandas as pd
```

# Preprocessing Data

```python
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()
for col in data:
    if data[col].dtype == 'object':
        data[col] = encode.fit_transform(data[col])
data

#Mengubah nilai-nilai string dalam kolom tipe object menjadi numerik
```

# Splitting Data

```python
from sklearn.model_selection import train_test_split

X = data.drop('HeartDisease', axis=1)
Y = data['HeartDisease']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
#untuk mengukur sejauh mana model dapat melakukan generalisasi pada data baru
```

# Modeling

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(x_train, y_train)
#digunakan untuk memprediksi kelas target
```

```python
y_pred = model.predict(x_test)
y_pred
#menghasilkan prediksi kelas target untuk setiap sampel dalam data pengujian menggunakan model yang telah terlatih.
```

# Accuracy Model

```python
from sklearn.metrics import accuracy_score

akurasi = accuracy_score(y_pred, y_test)
print(f'Akurasi Model adalah: {akurasi * 100}%')
#menampilkan akurasi model dengan persenan dari 1-100
```
